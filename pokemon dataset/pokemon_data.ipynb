# working with large datasets

```python
new_df = pd.DataFrame()
#new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified_pokemon.csv', chunksize=5):
    results = df.groupby(['Type 1' , 'Type 2']).count()
    
    new_df = pd.concat([new_df, results])

new_df
```
